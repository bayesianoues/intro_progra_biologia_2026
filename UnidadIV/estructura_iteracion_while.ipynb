# Bucle while en Python

A continuación, aprenderás sobre el bucle while en Python con la ayuda de ejemplos.

La iteración es un concepto fundamental en todos los lenguajes de programación, no solo en Python. Se utiliza para repetir un bloque de código múltiples veces según nuestros requerimientos.

El bucle while en Python se utiliza para iterar sobre un bloque de código mientras la condición de prueba se evalúe como True.

## Sintaxis del bucle while en Python

La sintaxis del bucle while en Python es:

```python
while test_condition:
  statements
```

Aquí, las instrucciones dentro del bucle while se ejecutan mientras .test_condition sea True.

Primero, se evalúa test_condition y, si es True, entonces se entra en el cuerpo del bucle. Después de una iteración, test_condition se evalúa nuevamente. Este proceso continúa hasta que test_condition se evalúa como False.

## Ejemplo 1: Bucle while en Python



````python
print('Numeros del 1 al 5:')
n = 1
while n <= 5:
  print(n)
  n = n + 1
````

Aquí es como el anterior programa trabaja:


| Iteración | Variable    | Condición: n | Acción                                     |
|-----------|------------|--------------|--------------------------------------------|
| 1.ª       | \( n = 1 \) (inicializada) | verdadero    | Se imprime 1. \( n \) aumenta a 2.        |
| 2.ª       | \( n = 2 \)    | verdadero    | Se imprime 2. \( n \) aumenta a 3.        |
| 3.ª       | \( n = 3 \)    | verdadero    | Se imprime 3. \( n \) aumenta a 4.        |
| 4.ª       | \( n = 4 \)    | verdadero    | Se imprime 4. \( n \) aumenta a 5.        |
| 5.ª       | \( n = 5 \)    | verdadero    | Se imprime 5. \( n \) aumenta a 6.        |
| 6.ª       | \( n = 6 \)    | falso        | El bucle termina.                          |

## Ejemplo 2: Bucle while en Python

````python
sum = 0

# Pida un número al usuario
n = int(input('Digite un número: '))

# Siga en el bucle hasta que n no sea igual a 0
while n != 0:

    # agregar el valor de n a la suma
    sum = sum + n

    # Pida un número al usuario
    n = int(input('Digite un número: '))

# Imprima la suma
print('Resultado :', sum)
````

Aquí es como el anterior programa trabaja:

| Iteración | Variables    | Condición: n != 0 | Acción                                        |
|-----------|--------------|-------------------|-----------------------------------------------|
| 1.ª       | suma = 0     | (3 != 0) es decir, Verdadero | suma = 0 + 3                            |
|           | n = 3        |                   | Se imprime "Ingrese un número"                |
| 2.ª       | suma = 3     | (1 != 0) es decir, Verdadero | suma = 3 + 1                            |
|           | n = 1        |                   | Se imprime "Ingrese un número"                |
| 3.ª       | suma = 4     | (1 != 0) es decir, Verdadero | suma = 4 + 1                            |
|           | n = 1        |                   | Se imprime "Ingrese un número"                |
| 4.ª       | suma = 5     | (0 != 0) es decir, Falso    | El bucle termina.                             |
|           | n = 0        |                   | Resultado: se imprime 5                       |


# Bucle while infinito

Si la condición de un bucle while es siempre Verdadera, el bucle se ejecuta un número infinito de veces (hasta que la memoria se llene). Esto se denomina bucle while infinito. Por ejemplo,

````python
while (True):
  print('Bucle infinito')
````

# Instrucción break en Python

La instrucción break se utiliza para terminar o finalizar el ciclo por completo. El control del programa pasa a la instrucción que está inmediatamente después del cuerpo del ciclo.

Veamos un ejemplo de esto:

## Ejemplo 1: Instrucción break en un bucle while

````python
n = 1

while True:
  print(n)
  n = n + 1
  break

print('Esto esta afuera del bucle while.')
````

Veamos otro ejemplo.

## Ejemplo 2: Instrucción break en un bucle while

````python
# El bucle termina cuando el item es igual a 5
item = 1

while item < 7:
    if item == 5:
        break
    print(item)
    item += 1

print('Fuera del bucle')
````

En este programa, iteramos a través de los números comenzando desde 1 hasta que el valor de item llega a 6. En cada iteración verificamos si item == 5; si esta condición es verdadera, entonces el ciclo se detiene con la instrucción break. De lo contrario, el valor de item se imprime y luego se incrementa en 1 para continuar con la siguiente iteración.

En la salida puedes ver que se imprimen los números del 1 al 4. Sin embargo, el resto de la secuencia no se muestra.

Esto se debe a que el bucle termina cuando item == 5, lo que hace que el control del programa salga inmediatamente del ciclo. Después de que el bucle finaliza, se imprime la instrucción Fuera del bucle.

# Instrucción continue de Python

La instrucción continue en Python omite el resto del código dentro del bucle para la iteración actual.

El bucle no termina, sino que continúa con la siguiente iteración.


## Ejemplo de instrucción continue

````python
for item in range(1, 7):
# Salta la iteración actual cuando el item es igual a 5
    if item == 5:
        continue
    print(item)
print('Fuera del bucle')
````

Este programa es el mismo que el ejemplo anterior, excepto que la instrucción break ha sido reemplazada por continue.

Aquí, puedes ver que la secuencia de números del 1 al 7 se imprime, pero se omite el 5 porque la instrucción continue omite el resto del código dentro de un bucle y continúa con la siguiente iteración.

Por lo tanto, el 5 se omite. Pero el número siguiente se imprime y, finalmente, se imprime la instrucción "Fuera del bucle".
