# Python for Loop

A continuación aprenderás sobre el bucle for en Python con la ayuda de ejemplos.

Un bucle for en Python nos permite acceder a cada elemento de una secuencia (string, lista, etc.) iterando sobre ella.

## Secuencias en Python

Una secuencia es una colección de elementos ordenados. Por ejemplo, una cadena de texto (string) es una secuencia de caracteres.



````python
text = 'Python'

numeros_pares = [2, 4, 6, 8]
````

Aquí hemos creado dos secuencias:

* text (string) — secuencia de caracteres. Para aprender más, visita Python String.

* numeros_pares (lista) — secuencia de números separados por comas y encerrados entre corchetes. Para aprender más, visita Python list.


## Python for Loop

Veamos la sintaxis para crear un bucle for en Python.

```python
for item in sequence:
    # code
```


Aquí, el código dentro del bucle for se ejecuta para cada elemento de la secuencia.

## Ejemplo: Python for Loop



````python
# create a sequence
text = "Python"

# access each character
for item in text:
    print(item)
````

En el código anterior, accedemos a cada carácter individual de la cadena usando un bucle for.

A continuación se explica cómo funciona el código:

| Iteración | Valor del elemento | Dentro del bucle      |
|-----------|--------------------|------------------------|
| Primera   | P                  | imprime el valor P     |
| Segunda   | y                  | imprime el valor y     |
| Tercera   | t                  | imprime el valor t     |
| Cuarta    | h                  | imprime el valor h     |
| Quinta    | o                  | imprime el valor o     |
| Sexta     | n                  | imprime el valor n     |

El bucle continúa hasta que alcanza el último elemento de la secuencia. Aquí, la cadena tiene 6 caracteres, por lo que el bucle se ejecuta 6 veces.




## Recorriendo una Lista

También podemos usar el bucle for para iterar a través de una lista. Por ejemplo,



````python
# crear una lista de numeros
numeros = [1, 2, 3, 4, 5]

# accesar cada numero
for num in numeros:
    print(num)
````

En el ejemplo anterior, hemos creado una lista de números y usamos un bucle for para acceder a cada número de la lista. Aquí está el funcionamiento detallado del código:


| Iteración | Valor de num | Dentro del bucle      |
|-----------|--------------|------------------------|
| Primera   | 1            | imprime el valor 1     |
| Segunda   | 2            | imprime el valor 2     |
| Tercera   | 3            | imprime el valor 3     |
| Cuarta    | 4            | imprime el valor 4     |
| Quinta    | 5            | imprime el valor 5     |

También podemos usar el bucle for para iterar a través de una lista de cadenas (strings).



````python
names = ['Jose', 'Maria', 'Pablo', 'Karen']

for name in names:
   print(name)
````

Aquí, en cada iteración del bucle, la variable name toma el valor de un elemento diferente de la lista names, y luego se imprime dicho valor. El bucle se ejecuta una vez por cada elemento en la lista.

## La sentencia break con el bucle for

Hasta ahora, nuestro bucle solo terminaba cuando se accedía a todos los elementos de una secuencia. Sin embargo, también podemos terminar el bucle usando la sentencia break.

La sentencia break dentro del bucle for finaliza el bucle inmediatamente. Por ejemplo:

````python
names = ['Jose', 'Maria', 'Pablo', 'Karen']

for name in names:

  if (name == 'Maria'):
    break

  print(name)
````

En el ejemplo anterior, hemos utilizado la sentencia break que se ejecuta si el valor de name es 'Maria'.

Cuando ejecutamos este código, solo obtenemos Jose como salida. Entendamos el funcionamiento de este bucle for.

| Iteración | Valor de name | ¿Condición name == 'María'? | Acción dentro del bucle          | ¿Se imprime? |
|-----------|---------------|------------------------------|-----------------------------------|--------------|
| Primera   | 'José'       | False                        | print(name) → Steve               | Sí           |
| Segunda   | 'Maria'       | True                         | Se ejecuta break (el bucle termina) | No           |

## La sentencia continue con el bucle for

Cuando usamos la sentencia continue con un bucle for, esta omite la iteración actual del bucle y comienza la siguiente iteración. Veamos un ejemplo.

````python
# crea una lista de cadenas

names = ['Jose', 'Maria', 'Pablo', 'Karen']
# accede a cada valor de la lista

for name in names:
  if (name == 'Maria'):
    continue

  print(name)
````

En el ejemplo anterior, hemos utilizado la sentencia continue que se ejecuta cuando el valor de name es 'Maria'.

Si observas la salida, puedes ver que el valor 'Maria' no se imprime. Veamos cómo funciona este bucle for en cada iteración:

| Iteración | Valor de name | ¿Condición name == 'María'? | Acción dentro del bucle                                      | ¿Se imprime? |
|-----------|---------------|------------------------------|---------------------------------------------------------------|--------------|
| Primera   | 'Jose'       | False                        | print(name) → Steve                                           | Sí           |
| Segunda   | 'Maria'       | True                         | Se ejecuta continue (salta a la siguiente iteración)         | No           |
| Tercera   | 'Pablo'        | False                        | print(name) → Ruby                                            | Sí           |
| Cuarta    | 'Karen'     | False                        | print(name) → Jackson                                         | Sí           |

## El bucle for con range()

Hasta ahora hemos usado el bucle for para acceder a cada elemento de una secuencia. Sin embargo, también podemos ejecutar el bucle un número determinado de veces utilizando la función range().

Veamos un ejemplo. Supongamos que queremos ejecutar nuestro bucle 5 veces.

````python
# Bucle for que se ejecuta 5 veces
for count in range(1, 6):
    print('Ejecutando bucle for')
````

Aquí, hemos utilizado el bucle para imprimir 'Ejecutando el bucle for' 5 veces usando la función range(1, 6). En este caso, la función range() creará una secuencia de números del 1 al 5 (el segundo valor, 6, es exclusivo, no se incluye).

Entonces, el bucle for se ejecuta para cada valor de la secuencia.

Podemos verificar esto imprimiendo el valor de count en su lugar.

````python
# Bucle for que se ejecuta 5 veces
for count in range(1, 6):
    print(count)
````

## Ejemplo: Tabla de Multiplicar

Creemos un ejemplo más. Esta vez usaremos el bucle for con range() para crear una tabla de multiplicar.

````python
number = int(input('Ingrese un número: '))

for count in range(1, 11):
    product = number * count
    print(number, '*', count, '=', product)
````

El código anterior puede crear la tabla de multiplicar de cualquier valor que se ingrese.

# Bucle for anidado en Python

Python nos permite incluir un bucle for dentro de otro, lo que se conoce como bucle for anidado. Por ejemplo:

````python
# Bucle for externo
for number in range(1, 3):

    # Bucle for interno
    for value in range(1, 6):
        product = number * value
        print(number, '*', value, '=', product)
````

En el ejemplo anterior, el bucle externo se ejecuta 2 veces (number = 1 y 2). En cada iteración del bucle externo, el bucle interno se ejecuta 5 veces (value = 1, 2, 3, 4 y 5).

Por lo tanto, en total el bucle se ejecuta 10 veces.

## La sentencia else con el bucle for

El bucle for de Python puede tener un bloque else opcional que se ejecuta una vez que la ejecución del bucle se ha completado. Por ejemplo:

````python
# Bucle for externo
for number in range(1, 3):

    # Bucle for interno
    for value in range(1, 6):
        product = number * value
        print(number, '*', value, '=', product)
    else:
        print('La table de multiplicar para', number, 'esta completada.')
````

Aquí, hemos añadido un bloque else al bucle for. Una vez que el bucle for interno se completa, se ejecuta el bloque else.

Por eso, obtenemos 'La tabla de multiplicar del 1 está completa' y 'La tabla de multiplicar del 2 está completa' en la salida.
