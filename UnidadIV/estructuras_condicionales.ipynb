# Estructuras condicionales

En este tutorial aprenderemos sobre las sentencias if...else en Python con la ayuda de ejemplos.

Cuando programamos, hay momentos en los que necesitamos tomar decisiones basadas en ciertos valores o condiciones.

En Python usamos la sentencia if para crear programas que pueden tomar decisiones.


## Sintaxis de la sentencia if

La sintaxis básica de la sentencia if es la siguiente:

```python
if test_condition:
statement(s)
```

Si test_condition es True, entonces se ejecuta el cuerpo del if.

Sin embargo, si test_condition es False, el cuerpo del if se omite y no se ejecuta.

## Ejemplo: Verificar si un estudiante aprobó el examen

Supongamos que eres un estudiante universitario y para aprobar un examen necesitas obtener 50 puntos o más.

Veamos un programa que verifica si aprobaste el examen.


````python
score = int(input("Digite un número: "))

if score >= 50:
  print("Felicidades, has pasado el exámen.")
````

## Usando otro if para manejar el caso contrario

Podemos agregar otra sentencia if para manejar el caso en el que el estudiante no aprueba el examen.

````python
score = int(input("Digite un número: "))

if score >= 50:
  print("Felicidades, has pasado el exámen.")
if score < 50:
  print("Lo siento, has fallado el exámen.")
````

## Sentencia if...else en Python

La sentencia if puede tener una cláusula opcional else que se ejecuta cuando test_condition es False.

La sintaxis de if..else es la siguiente:

```python

if test_condition:
  statement(s)
else:
  statement(s)

```

### Funcionamiento de if...else

Si test_condition se evalúa como True:

* El código dentro de if se ejecuta

* El código dentro de else se omite

Si test_condition se evalúa como False:

* El código dentro de else se ejecuta

* El código dentro de if se omite  

Ejemplo

Veamos un ejemplo práctico:

````python
score = 35

if score >= 50:
  print("Felicitaciones, has pasado el exámen.")
else:
  print("Lo siento, has fallado el exámen.")
````

En este caso, como la condición score >= 50 se evalúa como False, se ejecuta el bloque else.

## Sentencia if...elif...else en Python

La sentencia if también puede tener múltiples cláusulas elif para manejar más de dos casos de prueba.

La sintaxis de if...elif...else es la siguiente:

```python
if test_condition1:
  statements_1
elif test_condition2:
  statement_2
else:
  statements_3
```

## Funcionamiento de if...elif...else

* Si test_condition1 se evalúa como True, entonces se ejecuta el código statement_1.

* Sin embargo, si test_condition1 se evalúa como False, entonces se evalúa test_condition2.

* Si test_condition2 es True, se ejecuta statement_2.

* Pero si test_condition2 es False, entonces se ejecuta statement_3.


## Ejemplo: Sentencia if...elif...else en Python



````python
score = 105

if score > 100 or score < 0:
  print("La nota es invalida.")
elif score >= 50:
  print("Felicitaciones, has pasado el exámen.")
else:
  print("Lo lamento, has fallado el exámen.")
````

**Salida**

```python
Score is invalid.
```

En este caso, como la condición score > 100 or score < 0 se evalúa como True, se ejecuta la sentencia dentro del bloque if.


## Sentencia if anidada (Nested if) en Python

En Python también podemos utilizar una sentencia if dentro de otra sentencia if.
Esto se conoce como una sentencia if anidada (nested if statement).

Sintaxis de una sentencia if anidada:

```python
if test_condition1:
  statement_1

  if test_condition2:
    statement_2
```


Nota

* Podemos anidar múltiples niveles de sentencias if.

* Podemos agregar else y elif a la sentencia if interna.

* También podemos colocar una sentencia if interna dentro de un bloque else o elif externo, si es necesario.
