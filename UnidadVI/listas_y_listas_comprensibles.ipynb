## Creación de listas

Una lista se crea colocando elementos dentro de [], separados por comas. Por ejemplo,



````python
# Una lista con 3 elementos
numbers1 = [1, 2, -5]
# Lista que contiene valores duplicados
numbers2 = [1, 2, -5, 2]
# lista vacía
numbers3 = []
````

````python
print(numbers1)
````

````python
print(numbers2)
````

````python
print(numbers3)
````

En Python, una lista puede contener elementos de diferentes tipos. Por ejemplo:

````python
# Una lista con 3 elementos de diferentes tipos
random_list = [5, 2.5, 'Hola']
````

````python
print(random_list)
````

Una lista también puede contener otra lista como elemento. Por ejemplo:

````python
# Una lista con dos elementos: 5 y [4, 5, 6]
random_list = [5, [4, 5, 6]]
````

````python
print(random_list)
````

## Acceder a los elementos de una lista

Los elementos de una lista están ordenados, y podemos acceder a elementos individuales usando índices. Por ejemplo:

````python
cars = ['Toyota', 'Honda', 'Hyundai']
# acceder al primer elemento
print(cars[0]) # Toyota
````

````python
# acceder al tercer elemento
print(cars[2]) # Hyundai
````

En Python, la indexación comienza desde 0. Es decir,

* el índice del primer elemento es 0
* el índice del segundo elemento es 1
* y así sucesivamente

Por lo tanto, cars[0] devuelve el primer elemento de la lista cars, cars[1] devuelve el segundo elemento, y así sucesivamente.

## Indexación negativa

Las listas en Python también soportan indexación negativa.

* el índice del último elemento es -1
* el índice del penúltimo elemento es -2 y así sucesivamente

````python
# acceder al último elemento
print(cars[-1]) # Hyundai
````

````python
# acceder al tercer elemento desde el final
print(cars[-3]) # Toyota
````

Por cierto, si el índice especificado no existe en la lista, Python lanza la excepción IndexError.

````python
# intentando acceder al cuarto elemento
# sin embargo, la lista solo tiene tres elementos
print(cars[3])
````

## Segmentación (Slicing) de una lista

También es posible acceder a una sección de elementos de la lista, no solo a un elemento individual (usando la notación de slicing). Por ejemplo:

````python
languages = ['Python', 'JavaScript', 'C++', 'Kotlin']
# primeros tres elementos -> 0, 1, 2
print(languages[0: 3]) # ['Python', 'JavaScript', 'C++']
# desde el segundo hasta el último elemento -> 1, 2, 3
print(languages[1: 4]) # ['JavaScript', 'C++', 'Kotlin']
````

Una cosa importante a recordar sobre el slicing de listas es que el índice inicial es inclusivo, mientras que el índice final es exclusivo.

Por lo tanto,

* languages[0: 3] devuelve una lista con los elementos en los índices 0, 1 y 2, pero no el 3.
* languages[1: 4] devuelve una lista con los elementos en los índices 1, 2 y 3, pero no el 4.

## Slicing: omitir índice inicial y final

Si usamos un índice inicial vacío, el slicing comienza desde el inicio de la lista. De manera similar, si usamos un índice final vacío, el slicing termina en el último índice.

````python
languages = ['Python', 'JavaScript', 'C++', 'Kotlin']
# elementos desde el índice 0 hasta el 1
print(languages[:2]) # ['Python', 'JavaScript']
# elementos desde el índice 2 hasta el último índice
print(languages[2:]) # ['C++', 'Kotlin']
````

## Modificar elementos de una lista

Las listas en Python son mutables. Es decir, pueden cambiarse. Y podemos modificar los elementos de una lista asignando nuevos valores. Por ejemplo:

````python
cars
````

````python
# cambiando el segundo elemento a 'Ford'
cars[1] = 'Ford'
print(cars)
````

Recuerda que la indexación de listas comienza en 0, no en 1. Por lo tanto, para cambiar el segundo elemento, debemos asignar un nuevo valor a cars[1] y no a cars[2].

## Eliminar elementos de una lista

Podemos eliminar uno o más elementos de una lista usando la sentencia del de Python. La sentencia del también puede usarse para eliminar la lista completa.

````python
cars = ['BMW', 'Tesla', 'Ford', 'Toyota']
# eliminando el segundo elemento
del cars[1]
print(cars) # ['BMW', 'Ford', 'Toyota']
````

````python
# eliminando el último elemento
del cars[-1]
print(cars) # ['BMW', 'Ford']
````

````python
# eliminando la lista
del cars
print(cars) # Error: name 'cars' is not defined
````

## Longitud de una lista en Python

Podemos encontrar la longitud de una lista usando la función incorporada len(). Por ejemplo:

````python
cars = ['BMW', 'Tesla', 'Ford', 'Toyota']
print(len(cars)) # 4
````

````python
del cars[0]
print(len(cars)) # 3
````

## Iterar a través de una lista

Es muy fácil iterar a través de una lista usando un bucle for.

````python
cars = ['BMW', 'Tesla', 'Ford', 'Toyota']
# iterando a través de la lista

for car in cars:
  print(car)
````

## Verificar si un elemento existe en la lista

Podemos usar la palabra clave in para verificar si un elemento existe en la lista o no.

````python
cars = ['BMW', 'Tesla', 'Ford', 'Toyota']
print('Kia' in cars) # False
````

````python
print('Kia' not in cars) # True
````

````python
print('BMW' in cars) # True
````

## Comprensión de listas en Python

La comprensión de listas es una forma concisa y elegante de crear listas. Por ejemplo:

````python
# este código significa
# crear una lista de x*x donde
# x toma valores de 1 a 5 (range(1, 6))
numbers = [x*x for x in range(1, 6)]
print(numbers) # [1, 4, 9, 16, 25]
````

Aquí, este código

```
numbers = [x*x for x in range(1, 6)]
```

es equivalente a

numbers = []

```
for x in range(1, 6):
   numbers.append(x * x)
```
