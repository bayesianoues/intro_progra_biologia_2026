## Soporte incorporado en Python

Python tiene dos métodos principales para ordenar una lista: sort() y sorted().


## Ordenar una lista en Python

El ordenamiento es una operación fundamental en programación, que permite organizar datos en un orden específico. Aquí hay un fragmento de código para darte una idea.

````python
# Inicializando una lista
a = [5, 1, 5, 6]

# sort modifica la lista dada
a.sort()
print(a)
````

````python
b = [5, 2, 9, 6]

# sorted no modifica la lista dada
# y retorna una nueva lista ordenada
bs = sorted(b)
print(b)
print(bs)
````

## Métodos de ordenamiento en Python

### Método sort() en Python

El método sort() ordena la lista en el mismo lugar (in-place) y modifica la lista original. Por defecto, ordena en orden ascendente.

````python
# Inicializando una lista
a = [5, 2, 9, 1, 5, 6]

# Ordenando la lista en orden ascendente
a.sort()
print("Lista ordenada (ascendente):", a)

a.sort(reverse=True)
print("Lista ordenada (descendente):", a)
````

### Función sorted() en Python

La función sorted() devuelve una lista ordenada. No solo funciona con listas, sino con cualquier iterable (lista, tupla, cadena, etc.). No modifica el contenedor original y devuelve una versión ordenada.

````python
# Inicializando una lista
a = [5, 2, 9, 1, 5, 6]

# Ordenando la lista en orden descendente
sa = sorted(a)
print("Lista ordenada (ascendente):", sa)

sa = sorted(a, reverse=True)
print("Lista ordenada (descendente):", sa)
````

Ordenamiento de otros contenedores:

* Ordenar una tupla: sorted() convierte la tupla en una lista ordenada de sus elementos.

* Ordenar una cadena: sorted() ordena los caracteres y devuelve una lista de caracteres ordenados.

````python
# Ordenando una tupla
a = (10, 12, 5, 1)
print(sorted(a))
````

````python
# Ordenando una cadena
st = 'gfg'
print(sorted(st))
````

## Implementación de diferentes algoritmos en Python

1. Bubble Sort

Bubble Sort es el algoritmo de ordenamiento más simple que funciona intercambiando repetidamente elementos adyacentes si están en el orden incorrecto.

* Compara elementos adyacentes y los intercambia si están desordenados.

* Itera múltiples veces, moviendo el elemento más grande al final en cada pasada.

* Incluye una optimización: si no hay intercambios, el arreglo ya está ordenado.

Cómo funciona Bubble Sort

* Comparar cada par de elementos adyacentes.
Si el primero es mayor que el segundo, intercambiarlos.

* Después de cada pasada, el mayor elemento “sube” al final.

* Repetir para el resto de la lista.

* Se detiene si no hay intercambios.

### Implementación en Python

````python
def bubbleSort(arr):
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
            #print(f"Iteracion i:{i}, j:{j}, {arr}")
        if not swapped:
            break

arr = [64, 34, 25, 12, 22, 11, 90]
bubbleSort(arr)
print(arr)
````

Explicación:

* for i in range(n): Ejecuta múltiples pasadas; cada pasada coloca el siguiente elemento más grande al final.

* swapped = False: Indicador (flag) para detectar si ocurre algún intercambio en la pasada actual.

* for j in range(0, n - i - 1): Itera sobre los elementos no ordenados; el rango se reduce a medida que el final queda ordenado.

* if arr[j] > arr[j + 1]: Verifica si hay una inversión entre elementos adyacentes.

* arr[j], arr[j + 1] = arr[j + 1], arr[j]: Intercambia los elementos en el mismo arreglo usando desempaquetado de tuplas.

* if not swapped: break: Termina anticipadamente si no ocurre ningún intercambio, lo que indica que el arreglo ya está ordenado.

2. Selection Sort

Selection Sort es un algoritmo basado en comparaciones que selecciona repetidamente el elemento más pequeño y lo coloca en su posición correcta.

* Encuentra el mínimo y lo coloca al inicio.
* Luego el segundo mínimo, y así sucesivamente.

````python

def selectionSort(array, size):
    for ind in range(size - 1):
        min_index = ind

        for j in range(ind + 1, size):
            if array[j] < array[min_index]:
                min_index = j

        array[ind], array[min_index] = array[min_index], array[ind]
        #print(f"Iteracion ind:{ind}, j:{j}, {array}")

arr = [-2, 45, 0, 11, -9, 88, -97, -202, 747]
size = len(arr)
selectionSort(arr, size)

print(arr)
````

Explicación:

* for ind in range(size) (Outer loop): Itera sobre cada índice, tratándolo como el inicio de la parte no ordenada.

* min_index = ind: Asume que el índice actual contiene el elemento más pequeño.

* for j in range(ind + 1, size) (Inner loop): Encuentra el índice del elemento mínimo en la parte no ordenada del arreglo.

* if array[j] < array[min_index]: Actualiza min_index cada vez que se encuentra un elemento más pequeño.

* Intercambiar elementos: Intercambia el elemento más pequeño encontrado con el primer elemento de la sección no ordenada.

3. Insertion Sort

Insertion Sort inserta cada elemento en su posición correcta dentro de una parte ya ordenada.

Comienza desde el segundo elemento.
Inserta cada elemento en su lugar correcto.

````python
def insertionSort(arr):
    n = len(arr)

    if n <= 1:
        return
    for i in range(1, n):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
        print(f"Iteracion i:{i}, j:{j}, {arr}")



arr = [12, 11, 13, 5, -1]
insertionSort(arr)
print(arr)
````

Explicación:

* for i in range(1, n): Itera sobre el arreglo desde el segundo elemento.

* key = arr[i]: Almacena el elemento actual que será insertado en la parte ordenada.

* Bucle while interno: Desplaza los elementos mayores que key una posición hacia la derecha.

* arr[j + 1] = key: Inserta key en su posición correcta dentro de la parte ordenada.
