## Tupla en Python

A continuación aprenderás sobre las tuplas en Python y las operaciones con tuplas con la ayuda de ejemplos.

En Python, una tupla es similar a una lista con una diferencia importante. No podemos cambiar los elementos de una tupla una vez que ha sido asignada, mientras que sí podemos cambiar los elementos de una lista.

## Creación de tuplas

Una tupla se crea colocando elementos dentro de (), separados por comas. Por ejemplo:

```python
# una tupla con 3 elementos
numbers = (1, 2, -5)
print(numbers)
```

```python
# tupla que contiene valores duplicados
numbers = (1, 2, -5, 2)
print(numbers)
```

```python
# tupla vacía
numbers = ()
print(numbers)
```

En Python, una tupla puede contener elementos de diferentes tipos. Por ejemplo:

```python
# una tupla con 3 elementos de diferentes tipos
random_tuple = (5, 2.5, 'Hola')
print(random_tuple)
```

Una tupla también puede contener otra tupla y otros tipos de colecciones como elemento. Por ejemplo:

```python
# una tupla con dos elementos: 5 y (4, 5, 6)
random_tuple = (5, (4, 5, 6))
```

## Acceder a elementos de una tupla

Los elementos de una tupla están ordenados, y podemos acceder a elementos individuales usando índices. Por ejemplo:

```python
cars = ('BMW', 'Tesla', 'Ford')

# acceder al primer elemento
print(cars[0])    # BMW

# acceder al tercer elemento
print(cars[2])    # Ford
```

En Python, la indexación comienza desde 0. Es decir,

* el índice del primer elemento es 0
* el índice del segundo elemento es 1
* y así sucesivamente

Por lo tanto, cars[0] da el primer elemento de la tupla cars, cars[1] da el segundo elemento, y así sucesivamente.

## Indexación negativa

Las tuplas en Python también soportan indexación negativa.

* el índice del último elemento es -1
* el índice del penúltimo elemento es -2
* y así sucesivamente

```python
cars = ('BMW', 'Tesla', 'Ford')

# acceder al último elemento
print(cars[-1])    # Ford

# acceder al tercer elemento desde el final
print(cars[-3])    # BMW
```

Por cierto, si el índice especificado no existe en la tupla, Python lanza la excepción IndexError.

```python
cars = ('BMW', 'Tesla', 'Ford')

# intentando acceder al cuarto elemento
# sin embargo, la tupla solo tiene tres elementos
print(cars[3])
```

## Slicing de una tupla

También es posible acceder a una sección de elementos de la tupla, no solo a un elemento individual (usando la notación de slicing). Por ejemplo:

```python
languages = ('Python', 'JavaScript', 'C++', 'Kotlin')

# primeros tres elementos -> 0, 1, 2
print(languages[0: 3])    # ('Python', 'JavaScript', 'C++')
```

```python
# desde el segundo hasta el último -> 1, 2, 3
print(languages[1: 4])    # ('JavaScript', 'C++', 'Kotlin')
```

Una cosa importante a recordar sobre el slicing de tuplas es que el índice de inicio es inclusivo mientras que el índice final es exclusivo.

Por lo tanto,

* languages[0: 3] devuelve una tupla con los elementos en los índices 0, 1 y 2, pero no el 3.
* languages[1: 4] devuelve una tupla con los elementos en los índices 1, 2 y 3, pero no el 4.

## Slicing: omitir índice inicial y final

Si usamos un índice inicial vacío, el slicing comienza desde el inicio de la tupla. De manera similar, si usamos un índice final vacío, el slicing termina en el último índice.

```python
languages = ('Python', 'JavaScript', 'C++', 'Kotlin')

# elementos desde el índice 0 hasta el 1
print(languages[:2])    # ('Python', 'JavaScript')

# elementos desde el índice 2 hasta el último índice
print(languages[2:])    # ('C++', 'Kotlin')
```

## Cambiar elementos de una tupla

Las tuplas en Python son inmutables. Es decir, no podemos cambiar los elementos de una tupla una vez creada.

Si intentamos cambiar los elementos de una tupla, obtendremos un error.

```python
cars = ('BMW', 'Tesla', 'Ford')

# intentando cambiar el segundo elemento
# Error: el objeto 'tuple' no soporta asignación de elementos
cars[1] = 'Toyota'

```

```python
print(cars)
```

## Eliminar una tupla

No podemos eliminar elementos individuales de una tupla porque son inmutables. Sin embargo, podemos eliminar la tupla completa usando la instrucción del.

```python
cars = ('BMW', 'Tesla', 'Ford', 'Toyota')

# eliminando la tupla cars
del cars
```

```python
print(cars)    # Error: el nombre 'cars' no está definido
```

## Longitud de una tupla en Python

Podemos encontrar la longitud de una tupla usando la función incorporada len(). Por ejemplo:

```python
cars = ('BMW', 'Tesla', 'Ford', 'Toyota')
print(len(cars))    # 4
```

## Métodos de tuplas en Python

Dado que las tuplas son inmutables, solo existen dos métodos para tuplas.

| Método    | Descripción                                               |
|-----------|-----------------------------------------------------------|
| index()   | devuelve el índice del primer elemento coincidente       |
| count()   | devuelve el recuento del elemento especificado en la tupla |

## Iterar a través de una tupla

Es muy fácil iterar a través de una tupla usando el bucle for.

```python
cars = ('BMW', 'Tesla', 'Ford', 'Toyota')

# iterando a través de la tupla
for car in cars:
    print(car)
```

## Comprobar si un elemento existe en la tupla

Podemos usar la palabra clave in para verificar si un elemento existe en la tupla o no.

```python
cars = ('BMW', 'Tesla', 'Ford', 'Toyota')

print('Kia' in cars)    # False
```

```python
print('Kia' not in cars)    # True
```

```python
print('BMW' in cars)    # True
```

## Uso de las tuplas

Si estamos seguros de que los elementos de una colección no cambiarán, es mejor usar tuplas en lugar de listas.

Esto se debe a que si modificamos accidentalmente los elementos de una tupla, obtendremos un error. Esto ayuda a reducir errores (bugs) en nuestra aplicación.

## Resumen de las tuplas en Python

En Python, una tupla es:

* colección — puede contener múltiples elementos
* ordenada — los elementos de una tupla están en orden
* inmutable — los elementos de una tupla no pueden cambiarse

## Matrices usando listas en Python

En Python, no existe un tipo nativo llamado "matriz" como tal (a diferencia de otros lenguajes), pero podemos representarlas fácilmente usando listas de listas.

¿Qué es una matriz?

Una matriz es una estructura bidimensional (filas y columnas). En Python, podemos representarla así:

```python
# Matriz de 3x3
matriz = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

print(matriz)
```

* matriz[0] → primera fila

```python
matriz[0]
```

* matriz[1][2] → elemento en fila 1, columna 2 (valor: 6)

```python
print(matriz[1][2])
```

## Recorrer una matriz

```python
for fila in matriz:
    for elemento in fila:
        print(elemento)
```

## Ejemplo: Representar un Sudoku

Un Sudoku es un excelente ejemplo de matriz 9x9:

```python
sudoku = [
    [5, 3, 0, 0, 7, 0, 0, 0, 0],
    [6, 0, 0, 1, 9, 5, 0, 0, 0],
    [0, 9, 8, 0, 0, 0, 0, 6, 0],
    [8, 0, 0, 0, 6, 0, 0, 0, 3],
    [4, 0, 0, 8, 0, 3, 0, 0, 1],
    [7, 0, 0, 0, 2, 0, 0, 0, 6],
    [0, 6, 0, 0, 0, 0, 2, 8, 0],
    [0, 0, 0, 4, 1, 9, 0, 0, 5],
    [0, 0, 0, 0, 8, 0, 0, 7, 9]
]
```

* El número 0 representa una celda vacía.
* Puedes acceder a cualquier celda con sudoku[fila][columna]

## Modificar valores


```python
sudoku[0][2] = 4  # Cambiar una celda
```

## Mostrar la matriz de forma ordenada

```python
for fila in sudoku:
    print(fila)
```

O más formateado:

```python
for fila in sudoku:
    print(" ".join(str(x) for x in fila))
```

## Ejemplo: Representar una imagen en escala de grises

Una imagen en blanco y negro puede representarse como una matriz, donde cada número indica la intensidad del píxel:

* 0 → negro
* 255 → blanco
* valores intermedios → tonos de gris

## Ejemplo de imagen pequeña

```python
imagen_numero = [
    [0, 0, 255, 0, 0],
    [0, 255, 255, 0, 0],
    [0, 0, 255, 0, 0],
    [0, 0, 255, 0, 0],
    [255, 255, 255, 255, 255]
]
```

Cada posición representa un píxel.

```python
import matplotlib.pyplot as plt

plt.imshow(imagen_numero, cmap='gray')
plt.colorbar()  # opcional: muestra la escala de grises
plt.title("Imagen del número 1")
plt.show()
```
