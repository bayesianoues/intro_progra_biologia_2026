# Funciones en Python

A continuacion, aprenderás sobre las funciones de Python con la ayuda de ejemplos.

Una función es un bloque de código que realiza una tarea específica. Son un componente fundamental en la programación en Python y son esenciales para la reutilización y organización del código.

Las funciones se utilizan para dividir programas grandes en componentes más pequeños y modulares, haciendo que el código sea más organizado, legible y mantenible.


# Declaración de Función

```{python}
def function_name(parameters):
    # code
```


Para crear una función, utilizamos la palabra clave def seguida del nombre de la función, paréntesis (), y dos puntos :. El cuerpo de la función se especifica usando indentación.

Veamos un ejemplo:


````python
def greet():
  print('Hola')
  print('Como estas?')
````
