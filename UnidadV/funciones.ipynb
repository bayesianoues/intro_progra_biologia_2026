# Funciones en Python

A continuacion, aprenderás sobre las funciones de Python con la ayuda de ejemplos.

Una función es un bloque de código que realiza una tarea específica. Son un componente fundamental en la programación en Python y son esenciales para la reutilización y organización del código.

Las funciones se utilizan para dividir programas grandes en componentes más pequeños y modulares, haciendo que el código sea más organizado, legible y mantenible.


# Declaración de Función

```{python}
def function_name(parameters):
    # code
```


Para crear una función, utilizamos la palabra clave def seguida del nombre de la función, paréntesis (), y dos puntos :. El cuerpo de la función se especifica usando indentación.

Veamos un ejemplo:


````python
def greet():
  print('Hola')
  print('Como estas?')
````

Aquí hemos definido una función llamada greet().

Cuando ejecutamos el programa, no vemos ninguna salida. Esto se debe a que simplemente definir una función no hace nada. Para poner la función en acción, necesitamos llamarla.

# Llamando a una función

En el programa anterior, declaramos una función llamada greet(). Para usar esa función, necesitamos llamarla.

Llamamos a la función usando el nombre de la función:

````python
# llamando a la funcion greet()
greet()
````

En el codigo anterior greet() llama o invoca a la función greet().

# Múltiples llamadas a una función

Una ventaja de definir una función es que podemos llamarla cualquier número de veces.



````python
# llamando a la funcion greet tres veces.
greet()
greet()
greet()
````

Nota: Necesitamos definir la función antes de llamarla.

# Argumentos de función

Supongamos que queremos hacer nuestra función greet() un poco más personal.

En lugar de imprimir Hola, Como estas?, queremos imprimir algo como Hola Juan o cualquier nombre de la persona.

Para esto, podemos usar argumentos de función:

````python
def greet(name):
  print(f"Hola {name}")
  print('Como estas?')
````

````python
greet('Juan')
````

Los argumentos de función se pasan dentro de los paréntesis durante la llamada a la función. Luego se puede acceder a ellos usando el parámetro name en la definición de la función.

# Pasar múltiples argumentos

Si necesitamos pasar múltiples argumentos a una función, podemos separarlos por comas.

Vamos a crear una función para sumar dos números.

````python
def add_numbers(n1, n2):
  result = n1 + n2
  print(f"La suma es: {result}")

number1 = 5
number2 = 6
add_numbers(number1, number2)
````

Hemos pasado number1 y number2 como argumentos a la función add_numbers(). Estos argumentos se reciben como n1 y n2 una vez que se pasan a la función add_numbers().

# Valor de retorno de una función



````python
def add_numbers(n1, n2):
    result = n1 + n2
    print(f"La suma es: {result}")

add_numbers(5, 6)
````

La función add_numbers() imprime la suma de dos números. Sin embargo, es mejor simplemente calcular la suma dentro de la función e imprimir el resultado en otro lugar.

Podemos lograr esto utilizando la sentencia return.

````python
def add_numbers(n1, n2):
    result = n1 + n2
    return result

result = add_numbers(5, 6)
print(f"La suma es: {result}")
````
