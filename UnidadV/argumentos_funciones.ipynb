# Argumentos de funciones en Python: posicionales, por palabra clave y por defecto

En este tutorial aprenderemos diferentes formas de pasar argumentos a una función.

Existen distintas maneras de pasar argumentos a una función en Python. Son las siguientes:

* Argumentos posicionales
* Argumentos por defecto
* Argumentos por palabra clave

## Argumentos posicionales en Python

Los argumentos posicionales son el tipo más básico de argumentos que se pasan a una función.

Cuando llamas a una función y proporcionas valores para sus parámetros, esos valores se asignan a los parámetros según su posición u orden en la lista de parámetros de la función. Por ejemplo:

```python
def add_numbers(n1, n2):
  result = n1 + n2
  return result

result = add_numbers(5.4, 6.7)
print(result)
```

Hemos llamado a la función add_numbers() con dos argumentos.

* 5.4 - asignado al primer parámetro n1
* 6.7 - asignado al segundo parámetro n2

Estos argumentos se pasan según su posición. Por lo tanto, se llaman argumentos posicionales.

Si eliminamos el segundo argumento, se producirá un mensaje de error:

```python
def add_numbers(n1, n2):
  result = n1 + n2
  return result

result = add_numbers(5.4)

print(result)
```

El código anterior arroja un error.

Este mensaje de error nos indica que debemos pasar un segundo argumento para n2 durante la llamada a la función, ya que nuestra función add_numbers() recibe 2 argumentos.

## Argumentos por defecto en Python

Los argumentos de una función pueden tener valores por defecto en Python. Los argumentos por defecto son parámetros en la definición de una función que tienen un valor predefinido asignado.

Estos valores por defecto se utilizan cuando no se proporciona un argumento para ese parámetro durante la llamada a la función.

Podemos proporcionar un valor por defecto a un argumento utilizando el operador de asignación =. Por ejemplo:

```python
def add_numbers(n1 = 100, n2 = 1000):
  result = n1 + n2
  return result


result = add_numbers(5.4)
print(result)
```

Aquí, n1 toma el valor pasado en la llamada a la función. Como no hay un segundo argumento en la llamada a la función, n2 toma el valor por defecto de 1000.

También podemos ejecutar la función sin ningún argumento:

```python
def add_numbers(n1 = 100, n2 = 1000):
  sum = n1 + n2
  return sum

result = add_numbers()

print(result)
```

## Argumentos por palabra clave

En Python, los argumentos por palabra clave son una forma de pasar argumentos a una función utilizando explícitamente los nombres de los parámetros.

En lugar de depender del orden de los argumentos, los argumentos por palabra clave te permiten especificar el argumento junto con el nombre del parámetro correspondiente.

```python
def greet(name, message):
    print(f'Hola {name}')
    print(message)

# llamar a greet con argumentos posicionales
greet('Pablo', 'Como estas?')

#
greet(message='Howdy?', name='Jill')
```

```python
def greet(name, message):
    print(f'Hola {name}')
    print(message)

# llamar a greet con argumentos posicionales
greet('Pablo', 'Como estas?')

#
greet(message='Howdy?', name='Jill')
```

```python
def greet(name, message):
    print(f'Hola {name}')
    print(message)

# llamar a greet con argumentos posicionales
greet('Pablo', 'Como estas?')

# llamar a greet con argumentos por palabra clave
greet(message='Como estas?', name='Brenda')
```

Durante la segunda llamada a la función greet(message='Como estas?', name='Brenda'), pasamos los argumentos usando sus nombres de parámetro.

La cadena 'Como estas?' está asociada con el parámetro message, y la cadena 'Brenda' está asociada con el parámetro name. Cuando llamamos funciones de esta manera, el orden (posición) de los argumentos puede cambiarse.
