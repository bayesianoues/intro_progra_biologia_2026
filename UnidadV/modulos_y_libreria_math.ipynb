# Módulos en Python y libreria math

Un módulo es un archivo que contiene código de Python que podemos usar en nuestro programa. Existen varias funciones integradas como print(), input() y sum() que están disponibles para nosotros de forma inmediata.

Además de estas funciones, Python también tiene varias funciones que están definidas dentro de un módulo, las cuales podemos usar después de importarlas.

Usemos uno de estos módulos llamado math:

````python
import math

number = 25
result = math.sqrt(number)
print(result)

print(math.pi)
````

Importamos el módulo math usando la instrucción import. Una vez que importamos un módulo, podemos usar todo lo que contiene utilizando el operador punto.

```
result = math.sqrt(number)
```

Como puedes ver en el código anterior, usamos math.sqrt(number) para llamar a la función sqrt() del módulo math. Esta función calcula la raíz cuadrada de un número.

Además, el valor de la constante pi se imprime directamente en la pantalla.

```
print(math.pi)
```

## Renombrar módulos

Al importar un módulo, también podemos cambiarle el nombre según nuestras necesidades.

Renombremos el módulo math como m:

````python
import math as m

number = 25

# usamos m.sqrt() en lugar math.sqrt()

result = m.sqrt(number)

print(result)

# usamos m.pi en lugar de math.pi

print(m.pi)
````

Obtenemos el mismo resultado que antes.

## Sentencia from...import en Python

Cuando importamos un módulo como en nuestros ejemplos anteriores, todo lo que está dentro del módulo está disponible para nosotros.

Sin embargo, si solo necesitamos importar una definición específica como una función o una constante, podemos usar la sentencia from..import.

````python
from math import sqrt

num = sqrt(64)
print(num)
````

Aquí, solo se importa la función sqrt() del módulo math. Cuando usamos esta sintaxis, debemos usar directamente la función sqrt() en lugar de math.sqrt().

También podemos importar múltiples definiciones desde un módulo en una sola línea usando esta sintaxis:

```
from module_name import definition1, definition2, definition3, ...
```

Implementemos esto en nuestro jupyter notebook.

````python
from math import pi, sin, sqrt

value = sin(pi/2)
print(value)

num = sqrt(64)
print(num)
````

También podemos usar la sentencia from...import para importar todas las definiciones de un módulo usando *:

````python
from math import *

value = sin(pi/2)
print(value)

num = sqrt(64)
print(num)
````

Aquí, el asterisco * significa importar todo.

Nota: Importar todas las definiciones con el símbolo de asterisco * es una mala práctica de programación y deberías intentar evitarlo en tu código.


## La función dir()

El módulo math viene con muchas más funciones y constantes útiles.

Podemos listar todo lo que está definido dentro de un módulo usando la función dir():

````python
import math

print(dir(math))
````

Como puedes ver, hay muchas otras funciones matemáticas, trigonométricas y logarítmicas dentro del módulo math.

**Nota**: El archivo del módulo math es parte del paquete de instalación de Python. Por eso podemos importarlo directamente.
