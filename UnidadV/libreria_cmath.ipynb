# Introducción a los números complejos

Un **número complejo** es un número de la forma:

$$ z = a + bi $$


donde:

- $a$ es la **parte real**
- $b$ es la **parte imaginaria**
- $i$ es la unidad imaginaria, tal que:

$$ i^2 = -1 $$


## Ejemplo

$$ z = 3 + 4i $$

- Parte real: 3  
- Parte imaginaria: 4

## Números complejos en Python

En Python, los números complejos se representan usando `j` en lugar de `i`:


```python
z = 3 + 4j
print(z)
```

También puedes acceder a sus partes:

```python
print(z.real)  # parte real
print(z.imag)  # parte imaginaria
```

## Forma polar

Un número complejo también puede representarse como:

$$ z = r(\cos \theta + i \sin \theta) $$

donde:

- $r$ es el **módulo** (distancia al origen)  
- $\theta$ es el **argumento** (ángulo)  

El módulo se calcula como:

$$ r = \sqrt{a^2 + b^2} $$



## Librería cmath en Python

La librería cmath está diseñada para trabajar con números complejos (a diferencia de math, que es solo para números reales).A continuacion realizaremos calculos con la libreria cmath.

```python
import cmath
```

Módulo (valor absoluto)

```python
z = 3 + 4j
print(abs(z))
```

Argumento

```python
print(cmath.phase(z))
```

Conversión a forma polar

```python
r, theta = cmath.polar(z)
print(r, theta)
```

Conversión de forma polar a cartesiana

```python
z = cmath.rect(r, theta)
print(z)
```

Raíz cuadrada de números negativos

```python
print(cmath.sqrt(-1))  # resultado: 1j
```

Funciones trigonométricas

```python
z = 1 + 1j

print(cmath.sin(z))
print(cmath.cos(z))
```

Exponenciales y logaritmos

```python
print(cmath.exp(1j))   # e^(i)
print(cmath.log(1 + 1j))
```

## Ejemplo: Resolver $x^2 + 1 = 0$ usando `cmath`

Queremos resolver la ecuación:


$ x^2 + 1 = 0 $


Sabemos que:

$ x^2 = -1 $

Por lo tanto:


$ x = \pm \sqrt{-1} $

En números complejos:


$ x = \pm i $

A continuación resolvamos la ecuación $ x^2 + 1 = 0 $
usando Python.


```python
# resolver x^2 + 1 = 0
# x = ± sqrt(-1)

raiz1 = cmath.sqrt(-1)
raiz2 = -cmath.sqrt(-1)

print("Solución 1:", raiz1)
print("Solución 2:", raiz2)
```
